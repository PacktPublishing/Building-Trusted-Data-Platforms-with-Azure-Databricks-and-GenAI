```python
import dlt
from pyspark.sql.functions import col
```

```python
@dlt.view(
    name="electroniz_bronze_products_view",
    comment="Streaming view with selected product columns"
)
def electroniz_bronze_products_view():
    df = spark.readStream.table("electroniz_catalog.electroniz_bronze_schema.electroniz_bronze_products")
    return df.select(
        col("product_code"),
        col("product_name"),
        col("product_category"),
        col("updated_at")
    )
```

```python
dlt.create_streaming_table(
    name="electroniz_silver_products",
    comment="Silver products table",
    table_properties={"quality": "silver"}
)

dlt.apply_changes(
    target="electroniz_silver_products",
    source="electroniz_bronze_products_view",
    keys=["product_code", "product_name"],
    sequence_by=col("updated_at"),
    stored_as_scd_type=1
)
```
