```python
import requests
from datacontract.data_contract import DataContract

# Step 1: Download YAML from GitHub 
url = "https://raw.githubusercontent.com/Moulisaluja/Data-Engineering-with-Apache-Spark-Delta-Lake-and-Lakehouse/main/project/datacontract_valid.yaml"

response = requests.get(url)
response.raise_for_status()

yaml_path = "/tmp/datacontract_valid.yaml"
with open(yaml_path, "w") as f:
    f.write(response.text)

print("Downloaded successfully!")

# Step 2: Use the /tmp path to create a DataContract object
dc = DataContract(data_contract_file=yaml_path, spark=spark)

# Lint check
lint_result = dc.lint()
print("Lint Result:", lint_result.result)
print("\nValidation checks:")
for check in lint_result.checks:
    print(f"  [{check.result}] {check.name}: {check.reason}")

print("\n" + "="*50)
```

```python
test_result = dc.test()

print("Test Result:", test_result.result)
print("\nChecks:")
for check in test_result.checks:
    print(f"  [{check.result}] {check.name}: {check.reason}")
#To display the results in a table

import pandas as pd

rows = []
for check in test_result.checks:
    rows.append({
        "Check": check.name,
        "Result": str(check.result),
        "Reason": check.reason or ""
    })

df = pd.DataFrame(rows)
display(df)
```
